```python
import pandas as pd
import os
path=input('輸入路徑:')
VLRROBA_list=pd.DataFrame(columns=['A'])
for i in os.listdir(r"{}".format(path)):#組合所有excel
    df=pd.read_excel(r"{}\{}".format(path,i))
    VLRROBA_list=pd.concat([df,VLRROBA_list],axis=0)
VLRROBA_list=VLRROBA_list.drop(['A'], axis=1)
VLRROBA_list.to_excel(r'D:\合併完成excel.xlsx')
```

```python
df=VLRROBA_list.groupby(['YEAR','MONTH','time','PRODUCT_TYPE'])['SHIP_QTY','INSPECTION_QTY','Cost','YINSPECTION_QTY','YNG_QTY'].sum().reset_index()
```

```python
df['OBA/sorting rate']=df['INSPECTION_QTY']/df['SHIP_QTY']
df['OBA/sorting 不良率']=df['YNG_QTY']/df['YINSPECTION_QTY']
df.to_excel(r'D:\PD_Table.xlsx')
```
